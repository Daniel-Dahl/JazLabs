# Launch Servers, Clients, and Widgets

This notebook launches:
- Camera server/client/widget (First Light C-Red3_2Lite)
- digiHolo viewer
- Optical switch server/client/widget
- Pol switch server/client/widget
- Luminos stage server/client/widget

```python
# -----------------------------
# Edit these for your hardware
# -----------------------------
CAMERA_TYPE = "First Light C-Red3_2Lite"
CAMERA_KWARGS = {"CameraIdx": 0, "CalibrationFile": None}
CAM_HOST = "127.0.0.1"
CAM_COMMAND_PORT = 50731
CAM_FRAME_PUB_PORT = 50732

OPT_SWITCH_PORT = "COM12"
OPT_SWITCH_HOST = "127.0.0.1"
OPT_SWITCH_COMMAND_PORT = 50831

POL_SWITCH_PORT = "COM4"
POL_SWITCH_HOST = "127.0.0.1"
POL_SWITCH_COMMAND_PORT = 50841

LUMINOS_STAGE_PORT = "COM5"
STAGE_HOST = "127.0.0.1"
STAGE_COMMAND_PORT = 50931
```

```python
import JazLabs.hardware.Cameras.Camera_Server as CameraZMQServerlib
import JazLabs.hardware.Cameras.Camera_Client as CameraClientlib
import JazLabs.hardware.Cameras.Camera_Widget as CameraWidgetlib
import JazLabs.hardware.Cameras.Camera_Viewer as CamViewerlib


import JazLabs.hardware.digHolo.digHolo_pylibs.digHolo_Veiwer  as digholoVeiwerlib


import JazLabs.hardware.Lasers.Anritsu.AnritsuMG963xLaser as Laserlib

import JazLabs.hardware.OpticalSwitch.OpticalSwitch_Server as OSServerlib 
import JazLabs.hardware.OpticalSwitch.OpticalSwitch_Client as OSClientlib
import  JazLabs.hardware.OpticalSwitch.OpticalSwitch_widget as OSWidgetlib
# launch_optical_switch_widget_process

import JazLabs.hardware.MotorisedStages.MotorisedStage_Server as MotorisedStageZMQServerlib
import JazLabs.hardware.MotorisedStages.MotorisedStage_Client as MotorisedStageClientlib
import JazLabs.hardware.MotorisedStages.MotorisedStage_Widget as MotorisedStage_Widgetlib 

import JazLabs.procedures.Multi_Instrument_Measurement_Routines.Multi_Instrument_Measurement_Routines as Measurementlib
```

```python
InitaliseLaser=1
if InitaliseLaser==1:
    # Laser=Laserlib.LaserObject_JDSUniphaseTunableLaser(LaserID="GPIB0::1::INSTR")
    Laser=Laserlib.LaserObject(port="COM3")

else:
    del Laser
```

```python
camera_server = CameraZMQServerlib.CameraZMQServer(
    host=CAM_HOST,
    command_port=CAM_COMMAND_PORT,
    frame_pub_port=CAM_FRAME_PUB_PORT,
    CameraType=CAMERA_TYPE,
    CameraKwargs=CAMERA_KWARGS,
)
camera_server.startProcess()

camera_client = CameraClientlib.CameraClient(
    host=CAM_HOST,
    command_port=CAM_COMMAND_PORT,
    frame_pub_port=CAM_FRAME_PUB_PORT,
    client_id="nb_camera_client",
)
viewer = CamViewerlib.CameraViewer(
    host=CAM_HOST,
    command_port=CAM_COMMAND_PORT,
    frame_pub_port=CAM_FRAME_PUB_PORT,
    initial_scale=1,
)
viewer.startProcess()

camera_widget = CameraWidgetlib.launch_camera_widget(
    host=CAM_HOST,
    command_port=CAM_COMMAND_PORT,
    frame_pub_port=CAM_FRAME_PUB_PORT,
)

print("Camera server/client/widget launched")
```

```python
digholo_viewer = digholoVeiwerlib.digholoWindow(
    host=CAM_HOST,
    command_port=CAM_COMMAND_PORT,
    frame_pub_port=CAM_FRAME_PUB_PORT,
)
print("digiHolo viewer initialised")
```

```python
opt_switch_server = OSServerlib.OpticalSwitchZMQServer(
    host=OPT_SWITCH_HOST,
    command_port=OPT_SWITCH_COMMAND_PORT,
    switch_type="JDS_SC",
    switch_kwargs={"port": OPT_SWITCH_PORT},
)
opt_switch_server.startProcess()

opt_switch_client = OSClientlib.OpticalSwitchClient(
    host=OPT_SWITCH_HOST,
    command_port=OPT_SWITCH_COMMAND_PORT,
    client_id="nb_opt_switch_client",
)

opt_switch_widget_proc = OSWidgetlib.launch_optical_switch_widget_process(
    host=OPT_SWITCH_HOST,
    command_port=OPT_SWITCH_COMMAND_PORT,
)

print("Optical switch server/client/widget launched")
```

```python
# Pol switch uses the same switch server/client/widget stack, on a separate command port
pol_switch_server = OSServerlib.OpticalSwitchZMQServer(
    host=POL_SWITCH_HOST,
    command_port=POL_SWITCH_COMMAND_PORT,
    switch_type="JDS_Pol",
    switch_kwargs={"port": POL_SWITCH_PORT},
)
pol_switch_server.startProcess()

pol_switch_client = OSClientlib.OpticalSwitchClient(
    host=POL_SWITCH_HOST,
    command_port=POL_SWITCH_COMMAND_PORT,
    client_id="nb_pol_switch_client",
)

pol_switch_widget_proc = OSWidgetlib.launch_optical_switch_widget_process(
    host=POL_SWITCH_HOST,
    command_port=POL_SWITCH_COMMAND_PORT,
)

print("Pol switch server/client/widget launched")
```

```python
stage_server = MotorisedStageZMQServerlib.MotorisedStageZMQServer(
    host=STAGE_HOST,
    command_port=STAGE_COMMAND_PORT,
    stage_type="Luminos",
    stage_kwargs={"port": LUMINOS_STAGE_PORT},
)
stage_server.startProcess()

stage_client = MotorisedStageClientlib.MotorisedStageClient(
    host=STAGE_HOST,
    command_port=STAGE_COMMAND_PORT,
    client_id="nb_stage_client",
)

stage_widget_proc = MotorisedStage_Widgetlib.launch_motorised_stage_widget_process(
    host=STAGE_HOST,
    command_port=STAGE_COMMAND_PORT,
)

print("Luminos stage server/client/widget launched")
```
